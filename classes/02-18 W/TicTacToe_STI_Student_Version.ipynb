# 1) Tic-Tac-Toe as a System
A system has:
- inputs
- states
-  rules

Write 2–3 sentences: What makes this a “system”?

Tic-Tac-Toe is a system because it has interconnected components — state (the board), inputs (player moves), and outputs (the displayed board and result) — that work together to produce behavior. Each part depends on others: the display reads state, moves mutate state, and win-checking inspects state. The components follow defined rules (invariants) that keep the system consistent.



# 2) Essential State vs Derived State

## Essential (Required) State
Essential state is information you **must store** because you cannot reconstruct it.

Question:
If the program stopped right now, what must be saved to resume the game?

Essential state:

- The board (a 9-element list of values).

Why is that enough?

 The board encodes every move ever made. Since X always goes first, you can derive whose turn it is by counting X's and O's — so you don't need to store the current player separately.


## Derived State (Convenience)
Derived state is information you can compute from essential state.
It often exists for convenience/readability.

List 3 derived values you might compute from the board:

1) Whose turn it is (x_count == o_count → X's turn)
2) Whether anyone has won (check rows/cols/diagonals)
3) Whether the board is full (no values 1–9 remaining)

Why might we still compute/store derived values?

 For readability and convenience — it's cleaner to call current_player(board) than re-derive it inline every time.

# 5) Human move vs Computer move (Decision Logic)

Choosing a move is **decision logic**, not a state change.

Why might we want *two* functions?

- `get_human_move(board)`
- `get_computer_move(board)`

Write 1–2 sentences:

We want two separate functions because humans provide input interactively (via input()), while the computer calculates a move algorithmically. Separating them keeps decision logic clean and makes it easy to swap in a smarter AI later without touching the rest of the game loop.



# 6) Invariants

## Structural invariants (shape/allowed values)
List 2 structural invariants:

1) The board always has exactly 9 elements
2) Each cell contains only 10 (X), -10 (O), or an integer 1–9 (open)


## Logical invariants (rules that must always hold)
List 3 logical invariants:

1) X's count is always equal to or one more than O's count (X goes first)
2) A cell that has been played cannot be overwritten
3) The game stops as soon as a winner is found or the board is full
